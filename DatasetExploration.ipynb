```python
# Let's test a simple support vector machine to see if it can work.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV

X = df.iloc[:, 1:].copy().to_numpy()
y = df.iloc[:, 0].copy().to_numpy()
```

```python
X = (X - np.average(X, axis=0)) / np.std(X, axis=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = SVC(C=1.0, kernel='linear') # linear seems to be doing better than rbf?
clf.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, clf.predict(X_test), normalize="true")
```
